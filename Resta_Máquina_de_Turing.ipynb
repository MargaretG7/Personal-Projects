```python
from IPython.display import clear_output, Image
import time
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import numpy as np

entrada = input('Ingrese la operación a realizar: ')
elementos = []

numeros = entrada.split("-")

def display_graph(array, j):
    time.sleep(1.5)
    clear_current_output()
    dot = Digraph()
    # Establecer dirección vertical del grafo
    dot.graph_attr['rankdir'] = 'TB'
    dot.graph_attr['nodesep'] = '0'

    # Agregar los nodos al gráfico
    for i, element in enumerate(array):
        if i == j:
          dot.node(str(i), label=element, shape='rectangle', style= 'filled' ,fillcolor = '#63F758')
        else:
          dot.node(str(i), label=element, shape='rectangle')

    # Visualizar el gráfico
    dot.render('array_vertical', format='png')
    return Image(filename='array_vertical.png')

def clear_current_output():
    clear_output(wait=True)

for i, numero in enumerate(numeros):
    numero = numero.strip()
    if numero.isdigit():
        elementos.extend(list('1' * int(numero)))
    if i < len(numeros) - 1:
        elementos.append('-')
elementos = ['#'] + elementos + ['#']

for i in range(1, len(elementos)):
      graph(elementos,i)
      if elementos[i] == '#':
          elementos[i] = '='
          elementos.append('#')
          graph(elementos,i)

cont, band = 0, len(elementos)
pos = elementos.index('=')-1

def forGraph1(i,j):
  for x in range (i, j):
    graph(elementos, x)

def forGraph2(i,j):
  for y in range (j-1, i-1,-1):
    graph(elementos, y)

while (band != 0):
      if elementos[pos-cont] == '1':
            elementos[pos-cont] = 'X'
            forGraph1(pos-cont, len(elementos))
            elementos.insert(len(elementos)-1,'1')
            graph(elementos, len(elementos)-2)
            forGraph2(pos-cont, len(elementos)-2)
            cont += 1;
      else: break;
      band -=1;

cont, band = 0, len(elementos)
ddr, val = 0, 2
pos2 = elementos.index('-')-1
forGraph2(pos2+1, pos2+2)

while (band != 0):
     if elementos[1] == '-':
        graph(elementos, 0)
     if elementos[pos2-cont] == '1':
        elementos[pos2-cont] = 'X'
        graph(elementos, pos2-cont)
        forGraph1(pos2-cont, len(elementos))
        cont += 1;
        if elementos[len(elementos)-2] == '1' and ddr == 0:
           graph(elementos,len(elementos)-2)
           elementos.pop(len(elementos)-2)
           forGraph2(pos2-cont, len(elementos)-1)
        elif elementos[len(elementos)-val] == '=':
           elementos.insert(len(elementos)-1,'1')
           graph(elementos,len(elementos)-2)
           forGraph2(pos2-cont, len(elementos)-1)
           ddr +=1; val +=1
     else:
        break;
     band -=1

if ddr == 0 and elementos[pos+2] == '1':
  forGraph1(1,pos+2)
  elementos[pos+2] = '-'
  graph(elementos,pos+2)
  forGraph1(pos+3,len(elementos))
  elementos.insert(len(elementos)-1,'1')
  forGraph2(pos+1,len(elementos)-1)

  for i in range(pos+1, 0, -1):
        if elementos[i] == 'X':
            elementos[i] = '1'
        graph(elementos,i)

else:
  for i in range(0, pos+2):
        if elementos[i] == 'X':
            elementos[i] = '1'
        graph(elementos,i)
  forGraph2(1,pos+1)

def graph(a, i):
  display_graph(a, i)
  image = Image(filename='array_vertical.png')
  display(image)
```
